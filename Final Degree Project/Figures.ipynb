```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl
plt.style.use('seaborn-v0_8-paper')
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams['figure.dpi']=400
plt.rc('axes', titlesize=15)
plt.rc('axes', labelsize=15)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=12)    # fontsize of the tick labels
plt.rc('ytick', labelsize=12)    # fontsize of the tick labels
plt.rc('legend', fontsize=12)    # legend fontsize
```

```python
colours = ['#66c2a5','#fc8d62','#8da0cb','#e78ac3','#a6d854','C5','C6','C7','C8']

def language_shift(t,x):
	return (s0)*x**a0*(1-x)-(1-s0)*(1-x)**a0*x

def plot_shift(solution,label=None,x_line=False):
    plt.plot(solution.t, solution.y[0], label=label, c=color, linestyle='--', lw=2)
    if x_line==True and a0 !=1:
        plt.axhline(x_star, linestyle='dashdot', c=color)
    return

# Time span parameters
t_span = (0, 100)  # Simulate for t between 0 and 100
t_eval = np.linspace(*t_span, 500)  # Time points where we want the solution

# '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
# Initial condition
s0 = 0.6	# Prestige of language X
a0 = 1	# Volatility
x0 = [0.001,0.005,0.01,0.05,0.1]    # Initial proportion of population speaking the language
if a0 !=1:
	x_star = 1/(1+(s0/(1-s0))**(1/(a0-1)))


# Plot solutioins with diferent initial conditions
for x in x0:
    color = colours[x0.index(x)]
    solution = solve_ivp(language_shift, t_span, [x], t_eval=t_eval)
    plot_shift(solution,label=f'$x_0$ = {x}',x_line=True)

plt.xlabel('Time')
plt.ylabel('Proportion speaking language')
plt.title(f'Language Shift: s={s0}, $\\alpha={a0}$')
plt.legend(loc='lower right')
plt.ylim(-0.05,1.05)
plt.show()

```

```python
s0=0.5
x0_vec=(0.45,0.6)
a0=1.1
t = np.linspace(0,100,500)

# Time span parameters
t_span = (0, 100)  # Simulate for t between 0 and 100
t_eval = np.linspace(*t_span, 500)  # Time points where we want the solution

def perturb(t):
    return 1/(1+np.exp(-(np.log(x0/(1-x0)))*np.exp((a0-1)*t/2)))


f_t = np.vectorize(perturb)
for x0 in x0_vec:
    color = colours[x0_vec.index(x0)]
    plt.plot(t,f_t(t), label='',c=color, lw=3)
    solution = solve_ivp(language_shift, t_span, [x0], t_eval=t_eval, method='DOP853')
    plot_shift(solution,label='',x_line=False)
plt.axhline(y=0.5,xmin=0,xmax=100, linestyle='dashdot',c='black')
plt.xlabel('$t\:\:(c^{-1})$')
plt.ylabel('$x$')
# plt.legend()
plt.ylim(-0.05,1.05)

```
